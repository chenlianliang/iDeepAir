```python
#转换成27*27
import pandas as pd
Index_road=pd.read_csv(r'E:\data ming\accident_prediction\NYC_data\Experimental_data\index_road.csv')
Index_road=np.array(Index_road)
def transfer_729(A):
    B1=np.zeros([729])
    
    for i in range(354):
        B1[Index_road[i,0]]=A[i]
    B1=B1.reshape([27,27])
    return B1
#转换成354
def transfer_354(A):
    B1=np.zeros([354])
    A=np.array(A).reshape([729])
    for i in range(354):
        B1[i]=A[index_road[i]]
    return B1
```

```python
# GCN输出的一列向量转换成27 27
start=1000
Test_num=100
Acc_pred=np.load(r'GCN_results_07\task_1_18_9L_512_119.npy')
Acc_pre=Acc_pred.reshape([Test_num,354])
Acc_Pre27=np.zeros([Test_num,27,27])

for i in range(0,Test_num):
    Acc_Pre27[i,:,:]=transfer_729(Acc_pre[i])
    
    
#输入groundtruth的文件
Acc_T=np.load('Y_train_16p_mod.npy')
Acc_T=Acc_T[:,:,0]
Acc_T=Acc_T.reshape([Acc_T.shape[0],354])
Acc_T27=np.zeros([Acc_T.shape[0],27,27])
for i in range(0,len(Acc_T)):
    Acc_T27[i,:,:]=transfer_729(Acc_T[i])
```
